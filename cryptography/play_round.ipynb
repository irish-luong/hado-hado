```python
def search_less(arr, val, left, right):
    
    if arr[left] > val:
        return 0
    
    if right - 1 > left:

        mid = (right + left) // 2

        if arr[mid] <= val:
            left = mid
            return search_less(arr, val, left, right)

        if arr[mid] > val:
            right = mid
            return search_less(arr, val, left, right)
    
    
    if arr[right] == val:
        return right
        
    if arr[left] == val:
        return left
    
    if arr[right] < val:
        return right + 1
    
    if arr[left] < val:
        return left + 1

    return 0

def find_triplets(a, x):
    a = sorted(a)
    n = len(a)
    ans = 0

    for i in range(n-1, -1, -1):
        for j in range(i-1, -1, -1):
            
            remain = x - a[i] - a[j]
            the_rest = a[0:j]
            
            if not the_rest:
                continue

#             for k in range(len(the_rest) - 1, -1, -1):
#                 if the_rest[k] < remain:
#                     ans += len(the_rest[0:k+1])
#                     break
            print(remain)
            ans += search_less(the_rest, remain, 0, j-1)
    return ans
        
assert find_triplets([-2, 0, 1, 3], 2) == 2
# assert find_triplets([5, 1, 3, 4, 7], 12) == 4
# assert find_triplets([], 0) == 0

find_triplets([5, 1, 3, 4, 7], 12)
```

```python
def binary_search(arr, val, left, right):

    if right - 1 > left:
        mid = (right + left) // 2
        if arr[mid] <= val:
            left = mid
            return binary_search(arr, val, left, right)

        if arr[mid] > val:
            right = mid
            return binary_search(arr, val, left, right)
            
    if arr[left] == val:
        return left
    
    if arr[right] == val:
        return right
    return -1
    
binary_search([1,2,2,3,4], 4, 0, 4)     
```
